```python
import numpy as np
import pandas as pd

from scipy.optimize import curve_fit


import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('input_ginis.csv', skiprows=1,
                 names=['ISO3', 'year'] + ['SSP{}'.format(i) for i in range(1, 5)])
```

```python
df.head()
```

```python
isos = set(df[(df['SSP1'] < 15)]['ISO3'])
print(isos)
df.set_index(['ISO3', 'year']).sort_index().loc[list(isos)]['SSP1'].unstack(level=0).plot(legend=False)
```

```python
data = df.set_index(['ISO3', 'year']).loc['AUT']['SSP1']
data.plot()
```

```python
cross_idx = np.where(data < 15)[0][0]
cross_idx
```

```python
def coeffs(x, y, i, l=15, dx=5):
    df_i = (y[i + dx] - y[i - dx]) / (x[i + dx] - x[i - dx])
    b = -df_i / (y[i] + l)
    a = (y[i] + l) * np.exp(b * x[i])
    return a, b
```

```python
x = data.index.values
y = data.values
l = 0
i = cross_idx - 20
a, b = coeffs(x, y, i, l=l)
print(a, b)
f = lambda _: a * np.exp(-b * _) + l
```

```python
plt.plot(x, y, x[cross_idx - 30:], f(x[cross_idx - 30:]) - 30, x[cross_idx], y[cross_idx], 'go', x[cross_idx], f(x[cross_idx]), 'mo')
```
